```python
# Importing modules
import csv
import sys

#-----------------------------------------------------------------------------------------

# Declaring variables and file names
questions=[]
filename="Questions.csv"
fieldnames=["Question","A","B","C","correct_answer"]
#-----------------------------------------------------------------------------------------

#Reads contents of Questions.csv and loads it into memroy.
try:
  with open(filename, newline="",encoding="UTF-8") as f:
    reader=csv.DictReader(f)
    for row in reader:
      questions.append(row)
except FileNotFoundError:
  print("File not found.")
  with open(filename, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()   # creates file with headers only

#-----------------------------------------------------------------------------------------

#This function loads questions stored inside questions list back into the csv file, whenever there is a change in the questions list. The function is called at the following places: add, delete, edit, and at the time of exiting the programme.
def write_to_file():
  with open(filename, "w", newline="", encoding="UTF-8") as f:
    writer=csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(questions)
#-----------------------------------------------------------------------------------------

#MAIN Menu: This function let's you build a quiz or pay it.
def main_menu():
  while True:
    print("------Main Menu------")
    print("1. Play Quiz")
    print("2. Edit Questions")
    print("3. Exit")
    try:
      choice=int(input())
    except ValueError:
      print("Invalid input. Please enter a number.")
      continue
    if choice==1:
      play()
    elif choice==2:
      build_quiz()
    elif choice==3:
      write_to_file()
      return
    else:
      print("Enter a value between 1 and 3.")
#-----------------------------------------------------------------------------------------

#This function let's a user add/delete/display quiz questions.
def build_quiz():
  while True:
    print("--------Enter your choice--------")
    print("1. I want to delete a question.")
    print("2. I want to add a question.")
    print("3. I want to edit a question")
    print("4. I want to list all questions")
    print("5. I want to exit the quiz builder and return to MAIN menu.")
    try:
      edit_choice=int(input("Enter your choice:"))
    except ValueError:
      print("Invalid input. Please enter a number.")
      continue
    if edit_choice==1:
      del_question()
      continue
    elif edit_choice==2:
      add_question()
      continue
    elif edit_choice==3:
      edit_question()
      continue
    elif edit_choice==4:
      show_questions()
      continue
    elif edit_choice==5:
      write_to_file()
      return
    else:
      print("Invalid input. Please enter a number between 1 and 5.")
#-----------------------------------------------------------------------------------------------------------

#This function helps user navigate the quiz flow at the end of a response.
def navigate():
  while True:
    print("Do you want to cotinue(c)/exit(x)/Return to Main Menu(m)?")
    cont=input().lower()
    if cont=='c': return 'continue'
    elif cont=='x': return 'exit'
    elif cont=='m': return 'm'
    else: print("Invalid input. Please enter 'c', 'x', or 'm'.")
#------------------------------------------------------------------------------------------------------------

#This function displays all the quiz questions.
def show_questions():
  print("------LIST OF QUESTIONS------")
  for i in range(len(questions)):
    print(f"Question number:{i+1}: {questions[i]['Question']}")
    print(f"A. {questions[i]['A']}")
    print(f"B. {questions[i]['B']}")
    print(f"C. {questions[i]['C']}")
    print(f"Correct answer (Enter value from 1-3): {questions[i]['correct_answer']}")
#-----------------------------------------------------------------------------------------------------------

#This function let's the user add a new quiz question.
def add_question():
  while True:
    print("--------Add Question--------")
    print("Enter the question")
    Question=input()
    print("Enter option 1:")
    A=input()
    print("Enter option 2:")
    B=input()
    print("Enter option 3:")
    C=input()
    print("Enter the correct answer:")
    try:
      correct_answer=int(input())
    except:
      print("Please enter value between 1 and 3:")
      continue
    questions.append({"Question":Question, "A": A, "B": B, "C":C, "correct_answer":correct_answer})
    print("Question added successfully!")
    write_to_file()
    choice=navigate()
    if choice=='continue':
      continue
    elif choice=='exit':
      return
    elif choice=='m':
      return
    else:
      break

#-----------------------------------------------------------------------------------------------------------

#This function let's a user delete a quiz question.
def del_question(qdel=None):
  if qdel is None:
    while True:
      print("------DELETE A QUESTION------")
      try:
        qdel=int(input("Enter the question number that you want to delete:"))
      except ValueError:
        print("Incorrect value.")
      if 1<=qdel<=len(questions):
        del questions[qdel-1]
        print("Deletion successful")
      else:
        print(f"Enter value within 1 and {len(questions)}")
      choice=navigate()
      if choice=='continue':
        continue
      elif choice=='exit':
        return
      elif choice=='m':
        return
      else:
        break
  else:
    del questions[qdel-1]
    print("Deletion successful")
  write_to_file()
#-----------------------------------------------------------------------------------------------------------

#This function let's a user edit a quiz question. It asks the user for the question number to edit. It deletes it by calling delete function and then asks user to add a new function by calling the add() function.
def edit_question():
  print("------EDIT A QUESTION------")
  try:
    qedit=int(input("Enter the question number that you want to edit:"))
    print("Question will be removed and a new question will be added.")
    del_question(qedit)
    add_question()
  except ValueError:
    print("Incorrect value.")
    return questions
#-----------------------------------------------------------------------------------------------------------

#This function let's a user play the quiz.
def play():
  while True:
    print("GET.SET.GO!")
    print("When ready, press enter!")
    input()
    score=0
    for i in range(len(questions)):
      print(f"Q{i+1}. {questions[i]['Question']}")
      print(f"A. {questions[i]['A']}")
      print(f"B. {questions[i]['B']}")
      print(f"C. {questions[i]['C']}")
      answer=input("Enter your answer (Enter value from 1-3 or press space to pass):").strip()
      if answer.isdigit() and int(answer)==int(questions[i]['correct_answer']):
        score=score+1
      elif answer=="":
        continue
      else:
        score=score-1
    print(f"Your score is:{score}")
    choice=navigate()
    if choice=='continue':
      continue
    elif choice=='exit':
      return
    elif choice=='m':
      return
    else:
      break
#-------------------------------------------------------------------------------------------------------

#Main code block
main_menu()
```
